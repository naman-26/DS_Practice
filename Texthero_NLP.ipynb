```python
import pandas as pd
df= pd.read_csv("https://github.com/jbesomi/texthero/raw/master/dataset/bbcsport.csv")
df.head()
```

```python
## PCA
import texthero as hero

df['pca']=(
    df['text']
    .pipe(hero.clean)
    .pipe(hero.tfidf) ##vertorizing
    .pipe(hero.pca)

)
```

```python
hero.scatterplot(df, 'pca',  color='topic', title='PCA BBC Sport news')
```

# Kmeans

```python
df1 = pd.read_csv(
    "https://github.com/jbesomi/texthero/raw/master/dataset/bbcsport.csv"
)
```

```python
df1.head()
```

```python
df1['tfidf'] = (
    df1['text']
    .pipe(hero.clean)
    .pipe(hero.tfidf)
)
### Kmeans

df1['kmeans_labels'] = (
    df1['tfidf']
    .pipe(hero.kmeans, n_clusters=5)
    .astype(str)
)

df1['pca'] = df1['tfidf'].pipe(hero.pca)
```

```python
hero.scatterplot(df1, 'pca', color='kmeans_labels', title="K-means BBC Sport news")
```
