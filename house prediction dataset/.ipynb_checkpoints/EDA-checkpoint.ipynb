```python
## Data Analysis Phase
## Main aim is to understand more about data

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

## to display all the columns of dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
##printing the shape of dataset
print(dataset.shape)
```

```python
dataset.head()
```

# Missing Values

```python
## Here we will check the percentage of NAN values in each feature 
## Step 1 -- Make a list of featores which has missing values
features_with_na=[ features for features in dataset.columns 
                          if dataset[features].isnull().sum()>0
                 ]

##Step 2 -- Print the features name and the percentage of missing values
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), '% missing values')
    
```

### Since there are many missing values, we need to find the relationship between missing values and Sales price(dependent feature)

let's plot some diagrams for this relationship

```python
for feature in features_with_na:
    data =dataset.copy()
    
    # Let's make a variable that indicates 1 if the observation was missing or zero if not missing
    data[feature] = np.where(data[feature].isnull(),1 ,0)
    
    # Let's calculate the mean SalePrice, where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here, the relation between the missing values and the dependent variable is clearly visible. So we need to replace these NAN values with something meaningful which we will do in feature engineering.

From the above dataset, some features like id is not requried.

```python
print('Id of Houses {}'.format(len(dataset.Id)))
```

## Numerical Variables

```python
# List of numerical variables
numerical_features = [feature for feature in dataset.columns 
                                    if dataset[feature].dtypes !='O'
                      # O refers to object
                     ]
print('Number of numerical variables', len(numerical_features))

# visualize the numerical varibles
dataset[numerical_features].head()
```

### Temporal Variables(Eg: Datatime Variables)
From the dataset we have 4 year variables. We have to extract information from the dataetime variables like no. of days, no. of years. One example in above dataset can be the difference in the years between the year house was built and the year house was sold. We will be performing this analysis in the Feature Engineering.

```python
# List of variables that conatin year information
year_feature = [feature for feature in numerical_features 
                              if 'Yr' in feature or 'Year' in feature
               ]
year_feature
```

```python
# Let's explore the content of these year variables
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
## Let's analyze the temporal datetime variables
## We wil check whether there is relation between year the house is sold and Sales Price

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs YearSold')
```

```python
year_feature
```

```python
## Here we will compare the difference between all year feature with SalePrice

for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        ## We will capture the difference between year variables and year the house sold for
        data[feature]=data['YrSold']-data[feature]
    
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
## Numerical variable are usually of 2 types
##  Continous variable and Discrete variables

discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
## Let's find relationship between discrete feature and Sale Price

for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("Sale Price")
    plt.title(feature)
    plt.show()
```

```python
## there is a relationship between variable number and sale price
```

## Continuous Variable

```python
continuous_feature=[feature for feature in numerical_features if feature not in discrete_feature + year_feature+ ['Id']]
print("Continuous features Count {}".format(len(continuous_feature)))
```

```python
## Let's analyse the continuous values by creating histograms to understand the relationship between continuous feature and Sale Price
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
## we will be using logarithmic transformation

for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```
